```python
import requests
from bs4 import BeautifulSoup
import random
import pandas as pd
from tqdm import tqdm
from time import sleep
```

```python
url = 'https://nplus1.ru/' # сохраняем
page = requests.get(url) # загружаем страницу по ссылке
soup = BeautifulSoup(page.text, 'html')
urls = [link.get('href') 
        for link in soup.find_all('a') 
        if ('/news' in link.get('href') and not link.get('href').startswith('/news/'))]
```

```python
def GetNews(url0):
    """
    Returns a tuple with url0, author, description, title, final_text, rubrics, diff.
    Parameters:
    
    url0 is a link to the news (string)
    """
    page0 = requests.get(url0)
    soup0 = BeautifulSoup(page0.text, 'lxml')
    
    author = soup0.find_all('a', {'class' : 'underline'})[0].get_text()
    title = soup0.find_all('h1', {'class' : 'text-34 md:text-42 xl:text-52 break-words'})[0].get_text().replace('\xa0', ' ').strip()
    
    text = [t.text for t in soup0.find_all('div', {'class' : 'n1_material text-18'})]
    final_text = ' '.join(text)
    final_text = final_text.replace('\n', ' ')
    final_text = final_text.replace('\xa0', ' ')
    
    time, date, diff = [t.text for t 
                        in soup0.find_all('span', 
                                          {'class' : 'group-hover:text-main transition-colors duration-75'})[:3]]
    
    return url0, author, title, final_text, time, date, diff
```

```python
news = [] # это будет список из кортежей, в которых будут храниться данные по каждой новости

for link in tqdm(urls[:10]):
    try:
        res = GetNews(link)
        news.append(res)
        sleep(random.random())
    except:
        pass
```

```python
def get_urls():
    url = 'https://nplus1.ru/'  # сохраняем
    page = requests.get(url)  # загружаем страницу по ссылке
    soup = BeautifulSoup(page.text, 'html')
    urls = [link.get('href')
            for link in soup.find_all('a')
            if ('/news' in link.get('href') and not link.get('href').startswith('/news/'))]
    return urls

def GetNews(url0):
    """
    Returns a tuple with url0, author, description, title, final_text, rubrics, diff.
    Parameters:

    url0 is a link to the news (string)
    """
    page0 = requests.get(url0)
    soup0 = BeautifulSoup(page0.text, 'lxml')
    author = soup0.find_all('a', {'class': 'underline'})[0].get_text()
    title = soup0.find_all('h1',
                           {'class': 'text-34 md:text-42 xl:text-52 break-words'})[0]\
        .get_text()\
        .replace('\xa0', ' ')\
        .strip()
    text = [t.text for t in soup0.find_all('div', {'class': 'n1_material text-18'})]
    final_text = ' '.join(text)
    final_text = final_text.replace('\n', ' ')
    final_text = final_text.replace('\xa0', ' ')
#     time, date, diff = [t.text for t
#                         in soup0.find_all('span',
#                                           {'class': 'group-hover:text-main transition-colors duration-75'})[:3]]

    return url0, author, title, final_text

def get_news(data_list):
    news = []  # это будет список из кортежей, в которых будут храниться данные по каждой новости
    for link in data_list[:30]:
        try:
            res = GetNews(link)
            news.append(res)
#             sleep(random.random())
        except:
            pass

    df = pd.DataFrame(news)
    df.columns = ['link', 'author', 'header', 'text']
#     df['difficulty'] = df['difficulty'].apply(float)
    return df
```

```python
urls = get_urls()
df = get_news(urls)
```
